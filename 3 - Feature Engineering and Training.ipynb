```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import the file of contestant data
file_path = '/Users/ginagrainda/Desktop/survivor_data/survivor_contestants_encoded.csv'
data = pd.read_csv(file_path)
```

```python
print(data.head())
```

```python
correlation_matrix = data.corr(numeric_only=True)

# Extract the correlations with the target variable 'winner'
correlation_with_winner = correlation_matrix['winner'].sort_values(ascending=False)

# Select top 10 positively and 10 negatively correlated features (excluding 'winner')
top_positive_features = correlation_with_winner[1:11]  # Skip the first as it's 'winner' itself
top_negative_features = correlation_with_winner[-10:]

# Combine top positive and negative features
top_features = pd.concat([top_positive_features, top_negative_features])

# Check if the selected features exist in the dataset columns
selected_columns = top_features.index.tolist()
if 'winner' not in selected_columns:
    selected_columns.append('winner')  # Ensure target is included for correlations

# Create a heatmap data subset
heatmap_data = data[selected_columns].corr()

# Plot the heatmap
plt.figure(figsize=(14, 12))  # Increase figure size for better readability
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True)
plt.title('Correlation Heatmap for Selected Features', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
